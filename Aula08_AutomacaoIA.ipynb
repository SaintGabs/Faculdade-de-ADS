```python
'''
Missão 1

Crie um codigo em python simples
para verificar a temperatura de um sistema.
Se for maior do que 30 graus,
imprima um alerta de superaquecimento.
'''

def verificar_temperatura_sistema(temperatura):
    """
    Verifica a temperatura do sistema e imprime um alerta se for maior que 30°C.

    Args:
        temperatura (float): A temperatura atual do sistema em graus Celsius.
    """
    print("Temperatura atual do sistema: " + str(temperatura) + "°C")

    # Verifica se a temperatura é maior que 30
    if temperatura > 30:
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
    else:
        print("Temperatura do sistema está normal.")

# Solicita a temperatura ao usuário
temperatura_sistema_str = input("Digite a temperatura atual do sistema em °C: ")

try:
    temperatura_sistema = float(temperatura_sistema_str)
    verificar_temperatura_sistema(temperatura_sistema)
except ValueError:
    print("Entrada inválida. Por favor, digite um número para a temperatura.")
```

```python
'''
Missão 02

Crie um progama basico que orgaize arquivos bagunçados em pastas automaticamente
baseando apenas na extensão do arquivo(ex:pdf, jpg)
Utilize a bliblioteca OS ou shutil
'''
import os
import shutil

def organizar_arquivos(source_dir):
    """
    Organiza arquivos em pastas baseadas na extensão.

    Args:
        source_dir (str): O caminho do diretório contendo os arquivos bagunçados.
    """
    if not os.path.exists(source_dir):
        print("Diretório de origem '" + source_dir + "' não encontrado.")
        return

    print("\nIniciando organização dos arquivos em: " + source_dir)

    # Lista todos os itens no diretório de origem
    for filename in os.listdir(source_dir):
        # Constrói o caminho completo do arquivo
        file_path = os.path.join(source_dir, filename)

        # Verifica se é um arquivo (e não um diretório)
        if os.path.isfile(file_path):
            # Obtém o nome base e a extensão do arquivo
            base_name, extension = os.path.splitext(filename)

            # Remove o ponto da extensão e converte para minúsculas para usar como nome da pasta
            # Se não houver extensão (ex: 'Makefile'), usa 'sem_extensao'
            extension = extension[1:].lower() if extension else 'sem_extensao'

            # Define o diretório de destino para esta extensão
            destination_dir = os.path.join(source_dir, extension + '_files') # Ex: pdf_files, jpg_files

            # Cria o diretório de destino se ele não existir
            if not os.path.exists(destination_dir):
                os.makedirs(destination_dir)
                print("  Pasta criada: " + destination_dir)

            # Constrói o caminho completo do destino
            destination_path = os.path.join(destination_dir, filename)

            try:
                # Move o arquivo
                shutil.move(file_path, destination_path)
                print("  Movido: '" + filename + "' para '" + destination_dir + "'")
            except Exception as e:
                print("  Erro ao mover '" + filename + "': " + str(e))
        else:
            print("  Ignorando diretório/link: " + filename)

# --- Exemplo de uso e Teste ---
# Este código criará uma pasta 'arquivos_baguncados' no mesmo nível deste notebook,
# colocará alguns arquivos de exemplo nela e, em seguida, os organizará.

test_source_dir = "arquivos_baguncados"

# Remove a pasta de teste anterior e seus conteúdos para um teste limpo
if os.path.exists(test_source_dir):
    shutil.rmtree(test_source_dir)
    print("Pasta '" + test_source_dir + "' existente foi removida para um novo teste.")

# Cria a pasta de teste para arquivos bagunçados
os.makedirs(test_source_dir)

# Criar alguns arquivos de exemplo dentro dela para teste
with open(os.path.join(test_source_dir, "documento1.pdf"), "w") as f: f.write("conteudo pdf")
with open(os.path.join(test_source_dir, "imagem1.jpg"), "w") as f: f.write("conteudo jpg")
with open(os.path.join(test_source_dir, "relatorio.docx"), "w") as f: f.write("conteudo docx")
with open(os.path.join(test_source_dir, "foto.png"), "w") as f: f.write("conteudo png")
with open(os.path.join(test_source_dir, "script.py"), "w") as f: f.write("conteudo python")
with open(os.path.join(test_source_dir, "sem_extensao"), "w") as f: f.write("conteudo sem extensao")
with open(os.path.join(test_source_dir, "outro.pdf"), "w") as f: f.write("outro pdf")

print("\nPasta '" + test_source_dir + "' e arquivos de exemplo criados para teste.")

# Chame a função para organizar os arquivos
organizar_arquivos(test_source_dir)

print("\nOrganização concluída! Verifique a pasta 'arquivos_baguncados' para ver os resultados.")
```
