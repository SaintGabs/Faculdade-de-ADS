```python
#Soma de listas
numeros = [1, 2, 3, 4, 5] # Example: defining the list 'numeros'
soma = 0
i = 0
while i < len(numeros):
  soma = soma + numeros[i]
  i = i + 1
print(soma)
```

```python
#Numero primo
def primo(n):
  for i in range(2,n):
    if n % i == 0:
      return False
    return True
```

```python
#Imversão de String
def inverte_texto(texto):
  invertido = ''
  for letra in texto:
    invertido = letra + invertido
  return invertido
```
