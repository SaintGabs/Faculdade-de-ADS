```python
import math
def primo(n):
  if n <= 1:
    return False
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True
```

```python
class calculadora:
  def soma(self,a,b):
    return a+b

  def subtracao(self,a,b):
    return a-b

  def multiplicacao(self,a,b):
    return a * b

  def divisao(self,a,b):
    return a / b

# Crie uma instância da classe calculadora
calc = calculadora()

# Agora chame o método na instância
resultado = calc.divisao(10,2)
print(resultado)
```

```python
# Jogo da forca refatorado
def play_hangman():
    palavra = "python"
    letras_corretas = []
    letras_tentadas = []
    tentativas_max = 6
    tentativas_restantes = tentativas_max

    print("Bem-vindo ao Jogo da Forca!")

    while True:
        # Exibe o estado atual da palavra
        palavra_oculta = ""
        for letra_palavra in palavra:
            if letra_palavra in letras_corretas:
                palavra_oculta += letra_palavra
            else:
                palavra_oculta += "_"
        print(f"\nPalavra: {palavra_oculta}")
        print(f"Letras tentadas: {', '.join(sorted(letras_tentadas))}")
        print(f"Tentativas restantes: {tentativas_restantes}")

        # Verifica se o jogador ganhou
        if palavra_oculta == palavra:
            print("Parabéns! Você adivinhou a palavra!")
            break

        # Verifica se o jogador perdeu
        if tentativas_restantes == 0:
            print("\nSuas tentativas acabaram! Você perdeu.")
            print(f"A palavra era: {palavra}")
            break

        # Solicita uma letra ao jogador
        letra = input("Digite uma letra: ").lower()

        # Validação da entrada
        if not letra.isalpha() or len(letra) != 1:
            print("Por favor, digite apenas uma única letra.")
            continue

        if letra in letras_tentadas:
            print("Você já tentou essa letra. Tente outra.")
            continue

        letras_tentadas.append(letra)

        # Verifica se a letra está na palavra
        if letra in palavra:
            letras_corretas.append(letra)
            print("Acertou!")
        else:
            tentativas_restantes -= 1
            print("Errou!")

# Para jogar, descomente a linha abaixo:
play_hangman()
```

```python
import requests

def get_cep_data(cep):
    url = "https://brasilapi.com.br/api/cep/v2/"
    # key = ""  # Em APIs reais, você pode precisar de uma chave de autenticação

    try:
        # Monta a URL completa para a requisição GET
        endpoint = f"{url}{cep}"

        # Faz a requisição GET
        response = requests.get(endpoint)

        # Verifica se a requisição foi bem-sucedida (código de status 200)
        response.raise_for_status()

        # Converte a resposta JSON para um dicionário Python
        data = response.json()
        print(data)
        return data

    except requests.exceptions.HTTPError as http_err:
        print(f"Erro HTTP ao buscar CEP {cep}: {http_err}")
        return None
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Erro de Conexão ao buscar CEP {cep}: {conn_err}")
        return None
    except requests.exceptions.Timeout as timeout_err:
        print(f"Timeout da Requisição ao buscar CEP {cep}: {timeout_err}")
        return None
    except requests.exceptions.RequestException as req_err:
        print(f"Ocorreu um erro inesperado ao buscar CEP {cep}: {req_err}")
        return None

# Exemplo de uso da função:
cep_exemplo = "71693023"
cep_data = get_cep_data(cep_exemplo)
if cep_data:
    print(f"\nDados para o CEP {cep_exemplo}:")
    for k, v in cep_data.items():
        print(f"  {k}: {v}")
```
